```python
#2.Implement a program to read random line from a file containing lines of data in it(using random library)
#file_7 = open("U.txt",'x')
%%writefile -a U.txt.txt
Hello world.
How are you.
```

```python
import random
def random1(file):
    line = open(file).read().splitlines()
    return random.choice(line)

print(random1("U.txt.txt"))
```
